## Importing Libraries

```python
# Pake yang ini
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler


from fast_ml.model_development import train_valid_test_split
```

## Data Extraction

```python
TRAIN_PATH = r'../../Datasets/train.csv'
TEST_PATH = r'../../Datasets/test.csv'
SAMPLE_SUBMISSION_PATH = r"../../Datasets/sample_submission.csv"
```

```python
train_dat = pd.read_csv(TRAIN_PATH)
test_dat = pd.read_csv(TEST_PATH)
```

```python
train_dat.info()
```

## Data Prep


```python
#drop id
train_dat = train_dat.drop(columns='id')
test_dat = test_dat.drop(columns='id')
```

### Null Handling

```python
# fill null values with mean
train_dat = train_dat.fillna(train_dat.mean())
test_dat = test_dat.fillna(test_dat.mean())

train_dat.info()
```

### Encoding

#### One hot encoding

```python
# Perform one-hot encoding
train_dat = pd.get_dummies(train_dat, columns=['attribute_0', 'attribute_1'])
test_dat = pd.get_dummies(test_dat, columns=['attribute_0', 'attribute_1'])
```

```python
test_dat.info()
```

#### Label Encoding

```python
label_encoder = LabelEncoder()
train_dat['product_code_encoded'] = label_encoder.fit_transform(train_dat['product_code'])
train_dat =train_dat.drop(columns='product_code')
test_dat['product_code_encoded'] = label_encoder.fit_transform(test_dat['product_code'])
test_dat =test_dat.drop(columns='product_code')
```

```python
test_dat.head()
```

# Data Preprocessing

Feature Engineering

Outlier Handling

# SPLIT TRAIN AND TEST

```python
# Memisahkan fitur dengan target
full_train = train_dat.drop(columns=["failure"])
full_target = train_dat["failure"]
```

```python
# Train test split
X_train, X_test, y_train, y_test = train_test_split(
                                        full_train,
                                        full_target,
                                        test_size=0.3,
                                        random_state=42,
                                        )
```

# FEATURE SCALING

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(X_train.shape)
print(X_test.shape)
```

#  MODEL

```python
def metrics(y_true, y_pred):
    print("MSE  :", mean_squared_error(y_true, y_pred))
    print("RMSE :", mean_squared_error(y_true, y_pred)**(0.5))

def train_eval_models(models: dict, X_train, X_test, y_train, y_test):
    for model in models:
        m = model
        m.fit(X_train, y_train)
        y_pred = m.predict(X_test)
        print(model.__class__.__name__, models[model])
        metrics(y_test, y_pred)
```

### SVM, KNN, Naive Bayes

## SVM

```python
svm_model = SVC()
svm_model.fit(X_train, y_train)

y_pred = svm_model.predict(X_test)

metrics(y_test, y_pred)
```

## KNN

```python
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)

y_pred = knn_model.predict(X_test)

metrics(y_test, y_pred)
```

## Naive Bayes

```python
## Hyperparameter Optimization for Naive Bayes
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import GaussianNB
import numpy as np  # Make sure numpy is imported

# Define the parameter values that should be searched
var_smoothing = np.logspace(0,-9, num=100)

# Create a parameter grid: map the parameter names to the values that should be searched
param_grid = dict(var_smoothing=var_smoothing)

# Instantiate the Naive Bayes model
nb_model = GaussianNB()

# Instantiate the grid
grid = GridSearchCV(nb_model, param_grid, cv=10, scoring='accuracy')

# Fit the grid with data
grid.fit(X_train, y_train)

# View the complete results
grid.cv_results_

# Examine the best model
print(grid.best_score_)
print(grid.best_params_)
print(grid.best_estimator_)

## Naive Bayes
# Best hyperparameters obtained from hyperparameter tuning
best_params = grid.best_params_
# var_smoothing=best_params['var_smoothing']
nb_model = GaussianNB(var_smoothing=best_params['var_smoothing'])
nb_model.fit(X_train, y_train)

y_pred = nb_model.predict(X_test)

metrics(y_test, y_pred)
```
